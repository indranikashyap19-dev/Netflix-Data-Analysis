```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile

# Unzip the dataset if it's in an archive
with zipfile.ZipFile('archive.zip', 'r') as zip_ref:
    zip_ref.extractall('.')

#Load dataset
df=pd.read_csv('netflix_titles.csv')

#Show first rows
df.head()
```

```python
#Check missing values
df.isnull().sum()

#Fill missing values
df.fillna('Unknown',inplace=True)
```

```python
sns.countplot(x='type',data=df)
plt.title('Movies vs TV Shows')
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title('Top Genres')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title('Content Released Over Years')
plt.show()
```

-The majority of content on Netflix consists of Movies rather than TV Shows

-Drama is the most frequently occurring genre in the dataset
